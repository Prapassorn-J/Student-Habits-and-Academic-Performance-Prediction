###การจัดการค่าที่หายไป (Missing Values)

```python
missing_values = df.isnull().sum()
missing_values
```

เราพบค่าที่หายไป (Missing Value) ทั้งหมด 91 ค่า ที่ parental_education_level จึงได้ลบออกไปเนื่องจากการลบ 91 ค่านี้ไม่ได้ส่งผลต่อภาพรวมของข้อมูล เป็นการลบไปเพียง 9.1% ของ dataset

```python
df = df.dropna(subset=['parental_education_level'])

missing_values = df.isnull().sum()
missing_values
```

###การตรวจสอบข้อมูลซ้ำ (Duplicate Data)

```python
duplicate_count = df.duplicated().sum()
print(f"พบข้อมูลซ้ำทั้งหมด: {duplicate_count} แถว")

# หากมี ให้ลบออก
if duplicate_count > 0:
    df = df.drop_duplicates()
    print("ลบข้อมูลซ้ำเรียบร้อยแล้ว")
else:
    print("ไม่พบข้อมูลซ้ำในชุดข้อมูลนี้")
```

###การตรวจสอบและจัดการค่าผิดปกติ (Outliers)

ตรวจสอบ Outliers ด้วย Boxplot

```python
import seaborn as sns
import matplotlib.pyplot as plt

num_cols = [
    'age', 'study_hours_per_day', 'social_media_hours', 'netflix_hours',
    'attendance_percentage', 'sleep_hours', 'diet_quality', 'exercise_frequency',
    'internet_quality', 'mental_health_rating', 'exam_score'
]

plt.figure(figsize=(14, 8))
sns.boxplot(data=df[num_cols], orient='h', palette='pastel')
plt.title('Boxplot of Numerical Features')
plt.show()
```

###การแปลงชนิดข้อมูล (Data Type Conversion)

```python
categorical_cols = ['gender', 'part_time_job', 'diet_quality',
                    'parental_education_level', 'internet_quality',
                    'extracurricular_participation']

for col in categorical_cols:
    df[col] = df[col].astype('category')

df.dtypes
```

* การแปลง Categorical Columns เป็น category ช่วยประหยัดหน่วยความจำ

* เตรียมตัวแปรสำหรับการ Encoding ก่อนนำไปใช้ใน Machine Learning

###การเข้ารหัสข้อมูลเชิงหมวดหมู่ (Categorical Encoding)

เราใช้ One-Hot Encoding ซึ่งเหมาะกับการแปลงข้อมูลที่ไม่มีลำดับ (Nominal) กับข้อมูลที่เป็น category ซึ่งมีทั้งหมด 6 Columns ได้แก่ gender, part_time_job, diet_quality, parental_education_level, internet_quality, extracurricular_participation เพราะไม่มีลำดับสูงต่ำชัดเจน ซึ่งจะใช้เทคนิค One-Hot Encoding เพื่อแปลงให้เป็นคอลัมน์ใหม่ที่มีค่า 0 หรือ 1



```python
categorical_cols = [
    'gender',
    'part_time_job',
    'diet_quality',
    'parental_education_level',
    'internet_quality',
    'extracurricular_participation'
]
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=False)
bool_cols = df_encoded.select_dtypes(include='bool').columns
df_encoded[bool_cols] = df_encoded[bool_cols].astype(int)
df_encoded = df_encoded.drop(columns=['student_id'])
df_encoded.head()
```

**สร้าง Feature ใหม่ (Feature Engineering)**

จุดประสงค์ของส่วนนี้คือการ สร้าง Feature ที่เหมาะสมที่สุด
เพื่อเพิ่มประสิทธิภาพของโมเดล Machine Learning ในการทำนายคะแนนสอบของนักเรียน (exam_score)

วัตถุประสงค์ของการทำ Feature Engineering

2. เพื่อลดจำนวน Features ที่ไม่สำคัญ และลดความซับซ้อนของโมเดล

3. เพื่อเพิ่มความสามารถในการพยากรณ์ และตีความผลลัพธ์ของโมเดล

ในข้อมูลชุดนี้  Feature เดิมสะท้อนพฤติกรรมแต่ละด้านอยู่แล้ว
อย่างไรก็ตาม สามารถสร้าง Feature“เชิงสัมพันธ์” ที่อธิบายพฤติกรรมเชิงรวมได้ เช่น

| `study_sleep_ratio`  | `study_hours_per_day / (sleep_hours + 1e-6)` | อัตราส่วนระหว่างเวลาเรียนต่อการพักผ่อน |
| `is_parent_high_edu` | 1 ถ้าผู้ปกครองมีวุฒิ ≥ ป.ตรี                 | บ่งชี้ถึงสิ่งแวดล้อมการศึกษาที่ดี      |

####Feature Engineering – Creating New Features

2. study_sleep_ratio

```python
df_encoded['study_sleep_ratio'] = df_encoded['study_hours_per_day'] / (df_encoded['sleep_hours'] + 1e-6)
```

###การปรับขนาดข้อมูล (Feature Scaling)

การปรับสเกลข้อมูล (Scaling) มีความสำคัญโดยเฉพาะเมื่อใช้โมเดลที่อ่อนไหวต่อค่าตัวเลข เช่น Linear Regression
โดยจะใช้ StandardScaler เพื่อให้ค่าเฉลี่ย = 0 และส่วนเบี่ยงเบนมาตรฐาน = 1

```python
from sklearn.preprocessing import StandardScaler
numeric_cols = []
for col in df_encoded.columns:
    if col != 'target':  # ถ้ามี target
        # เลือก column ที่ไม่ใช่ค่า 0/1 ล้วน (เพราะคือ one-hot)
        if not set(df_encoded[col].unique()).issubset({0, 1}):
            numeric_cols.append(col)

# 2) สร้าง scaler
scaler = StandardScaler()

# 3) Fit + transform เฉพาะ numeric columns
df_encoded[numeric_cols] = scaler.fit_transform(df_encoded[numeric_cols])

df_encoded.head()
```

**การเลือก Feature ที่สำคัญ (Feature Selection)**

หลังจากสร้าง Feature ใหม่แล้ว จำเป็นต้องเลือกเฉพาะตัวแปรที่ส่งผลต่อการพยากรณ์จริง ๆ
โดยใช้เทคนิค SelectKBest ซึ่งจัดอยู่ในกลุ่ม Filter Methods เนื่องจากสามารถประมวลผลได้อย่างรวดเร็วและช่วยคัดตัวแปรที่มีความสัมพันธ์ชัดเจนกับ target ทำให้โมเดลทำงานง่ายขึ้นรวมถึงลดการเกิด overfitting

####วิธี SelectKBest (Statistical Test - f_regression)

```python
from sklearn.feature_selection import SelectKBest, f_regression

X = df_encoded.drop(['exam_score'], axis=1)
y = df_encoded['exam_score']


selector = SelectKBest(score_func=f_regression, k=10)
X_new = selector.fit_transform(X, y)

selected_features = X.columns[selector.get_support()]
scores = selector.scores_[selector.get_support()]

feature_scores = pd.DataFrame({'Feature': selected_features, 'Score': scores})
feature_scores = feature_scores.sort_values(by='Score', ascending=False)

plt.figure(figsize=(8,5))
sns.barplot(data=feature_scores, x='Score', y='Feature', palette='pastel')
plt.title('Top 10 Features (SelectKBest)')
plt.show()

print("Feature ที่ผ่านการคัดเลือกโดย SelectKBest:")
display(feature_scores)
```

# หลักการของแต่ละอัลกอริทึม

1.   **Multiple Linear Regression** หลักการ : Multiple Linear Regression เป็นโมเดลเชิงเส้นที่ใช้สมการเชิงเส้นตรงรวมของทุกตัวแปรอิสระ (features) เพื่อทำนายค่าตัวแปรตาม (target) โดยปรับค่าค่าสัมประสิทธิ์ (coefficients) ให้ผลรวมของความคลาดเคลื่อน (error) ต่ำที่สุด

2.   **Ridge Regression** หลักการ : Ridge Regression มีการเพิ่ม L2 Regularization เพื่อควบคุมขนาดของค่าสัมประสิทธิ์ ทำให้โมเดลมีความเสถียรมากขึ้น โดยเฉพาะเมื่อตัวแปรต้นมีความสัมพันธ์กันสูง จุดแข็งของ Ridge คือความสามารถในการลด overfitting ได้โดยไม่จำเป็นต้องตัดตัวแปรใดออก ส่งผลให้โมเดลรักษาข้อมูลสำคัญทั้งหมดไว้ได้

1.   **Losso Regression** หลักการ : เพิ่ม L1 Regularization  ซึ่งมีคุณสมบัติทำให้ค่าค่าสัมประสิทธิ์บางตัวกลายเป็นศูนย์ โมเดลจึงสามารถเลือกฟีเจอร์ที่สำคัญได้โดยอัตโนมัติ

2.   **Random Forest Regression** หลักการ : Random Forest Regression เป็นโมเดลแบบ tree-based โดยใช้การสร้างต้นไม้หลายต้น (ensemble) และนำผลลัพธ์มาหาค่าเฉลี่ย สามารถเรียนรู้ความสัมพันธ์ทั้งเชิงเส้นและไม่เชิงเส้น



# การวิเคราะห์ข้อดีข้อเสียของแต่ละอัลกอริทึม  
1. Multiple Linear Regression (MLR)  
**ข้อดี**
*   มีโครงสร้างโมเดลที่เรียบง่ายและสามารถประมวลผลได้อย่างรวดเร็ว
*   เหมาะสำหรับข้อมูลที่มีความสัมพันธ์เชิงเส้นระหว่างตัวแปร
*   ให้ความสามารถในการตีความสูง โดยสามารถวิเคราะห์ทิศทางและขนาดของผลกระทบจากแต่ละตัวแปรได้อย่างชัดเจน
*   ลดความเสี่ยงต่อการเกิด overfitting เมื่อจำนวนตัวแปรไม่มากและความสัมพันธ์เป็นเชิงเส้น

*   ไม่สามารถรองรับความสัมพันธ์ที่มีลักษณะไม่เป็นเชิงเส้น
*   มีความไวต่อ outlier ซึ่งอาจส่งผลต่อค่าสัมประสิทธิ์
*   ประสิทธิภาพอาจลดลงหากเกิด multicollinearity ระหว่างตัวแปรอิสระ

*   ลดโอกาสเกิด overfitting เมื่อตัวแปรมีจำนวนมากหรือตัวแปรบางส่วนมีความสัมพันธ์สูง
*   รักษาโครงสร้างของโมเดลเชิงเส้นไว้ ทำให้ตีความได้ง่ายกว่าโมเดลที่ซับซ้อนกว่า  
**ข้อจำกัด**
*   ไม่สามารถตัดตัวแปรออกจากโมเดล เนื่องจากค่าสัมประสิทธิ์ไม่ลดลงจนเป็นศูนย์
*   ยังคงเป็นโมเดลเชิงเส้น จึงไม่สามารถจับความสัมพันธ์ไม่เป็นเส้นตรงได้  

3. Lasso Regression  
**ข้อดี**
*   สามารถทำ Feature Selection อัตโนมัติ โดยดึงค่าสัมประสิทธิ์ของตัวแปรที่ไม่สำคัญให้เป็นศูนย์
*   ช่วยลดความซับซ้อนของโมเดลและทำให้ตีความได้ชัดเจนขึ้น
*   ลดความเสี่ยงของ overfitting เมื่อข้อมูลมีตัวแปรจำนวนมากหรือมีตัวแปรบางส่วนที่ไม่จำเป็นต่อการทำนาย

####ฟังก์ชันประเมินผลโมเดล (Evaluation Function) และ Plot กราฟ



```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def evaluate_regression_model(name, model, X_train, y_train, X_test, y_test, results_list):

    model.fit(X_train, y_train)

    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    mae_train = mean_absolute_error(y_train, y_train_pred)
    mae_test = mean_absolute_error(y_test, y_test_pred)
    rmse_train = np.sqrt(mean_squared_error(y_train, y_train_pred))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_test_pred))
    r2_train = r2_score(y_train, y_train_pred)
    r2_test = r2_score(y_test, y_test_pred)

    print(f"=== {name} ===")
    print(f"MAE (train): {mae_train:.3f}")
    print(f"MAE (test) : {mae_test:.3f}")
    print(f"RMSE(train): {rmse_train:.3f}")
    print(f"RMSE(test) : {rmse_test:.3f}")
    print(f"R² (train) : {r2_train:.3f}")
    print(f"R² (test)  : {r2_test:.3f}")
    print("-"*40)

    results_list.append({
        "Model": name,
        "MAE_train": mae_train,
        "MAE_test": mae_test,
        "RMSE_train": rmse_train,
        "RMSE_test": rmse_test,
        "R2_train": r2_train,
        "R2_test": r2_test
    })

    return y_test_pred

def plot_actual_vs_predicted(y_actual, y_predicted, model_name):
    """
    สร้าง scatter plot ของค่าจริง (Actual) vs. ค่าทำนาย (Predicted)
    """
    plt.figure(figsize=(8, 6))
    sns.set_style('whitegrid')

    sns.regplot(x=y_actual, y=y_predicted,
                scatter_kws={'alpha': 0.3, 's': 20, 'color': 'blue'},
                line_kws={'color': 'red', 'lw': 2})

    min_val = min(y_actual.min(), y_predicted.min())
    max_val = max(y_actual.max(), y_predicted.max())
    plt.plot([min_val, max_val], [min_val, max_val], 'k--', label='Ideal (Actual = Predicted)')

    plt.title(f'Actual vs. Predicted Values\n({model_name})', fontsize=16)
    plt.xlabel('Actual Value', fontsize=12)
    plt.ylabel('Predicted Value', fontsize=12)
    plt.legend()
    plt.grid(True)

    plt.show()
```

### **โมเดลที่ 1**  Multiple Linear Regression

```python
from sklearn.linear_model import LinearRegression

results = []

lin_reg = LinearRegression()

y_pred_lr = evaluate_regression_model(
    "Linear Regression",
    lin_reg,
    X_train, y_train,
    X_test, y_test,
    results
)

# 2. พล็อตกราฟของโมเดลนี้ทันที
plot_actual_vs_predicted(y_test, y_pred_lr, "Linear Regression")
```

การวิเคราะห์ : ค่าความแม่นยำ R²_test บนชุดทดสอบอยู่ที่ 0.893 ซึ่งสูงมาก แสดงให้เห็นว่าโมเดลสามารถอธิบายความแปรปรวนของข้อมูลได้ประมาณ 89.3% นอกจากนี้ ความแตกต่างระหว่างค่าบนชุดฝึกและชุดทดสอบมีค่าน้อยมาก สะท้อนว่าโมเดลไม่มีปัญหา overfitting

**เมตริกที่ใช้ในการประเมินโมเดล (Metric Selection)** ได้แก่ MAE RMSE และ R^2  
*   MAE → วัดขนาดความผิดพลาดเฉลี่ยโดยรวม ลดอิทธิพล outlier (ยิ่งมีค่าน้อยยิ่งดี)
*   RMSE → วัดความแม่นยำและลงโทษ (penalize) ความผิดพลาดใหญ่ (ยิ่งมีค่าน้อยยิ่งดี)
*   R² → ดูความสามารถในการอธิบายความแปรปรวนของโมเดล (ยิ่งมีค่ามากยิ่งดี)

  ในการประเมินประสิทธิภาพของการทำนายผลการเรียน เราเลือกใช้ MAE, RMSE และ R² เป็น Metric เนื่องจากแต่ละตัวสะท้อนมุมมองของโมเดล Regression ที่แตกต่างกัน MAE วัดขนาดความผิดพลาดเฉลี่ยของโมเดลโดยรวม ลดอิทธิพลของค่า outlier ทำให้เข้าใจความคลาดเคลื่อนทั่วไป RMSE วัดความแม่นยำและให้ค่าน้ำหนักกับความผิดพลาดที่สูงมาก ทำให้เห็นความผิดพลาดรุนแรงที่ MAE อาจละเลย ส่วน R² วัดความสามารถของโมเดลในการอธิบายความแปรปรวนของตัวแปรตาม ทำให้เห็นภาพรวมว่าข้อมูลถูกอธิบายโดยโมเดลได้ดีเพียงใด การใช้ทั้งสาม Metric ร่วมกันช่วยให้ประเมินโมเดลได้ครบทั้งขนาดความผิดพลาดเฉลี่ย ความผิดพลาดรุนแรง และความสามารถในการจับความสัมพันธ์ของข้อมูลได้

พบว่าโมเดลแบบ Multiple Linear Regression มีประสิทธิภาพดีที่สุดเมื่อทดสอบด้วยชุดข้อมูลใหม่ โดยให้**ค่า MAE และ RMSE ต่ำที่สุด** พร้อมค่า **R² สูงที่สุด** แสดงถึงความสามารถในการอธิบายความแปรปรวนของคะแนนได้ดี

นอกจากนี้ Multiple Linear Regression ยังมีข้อดีคือเป็นโมเดลที่ เรียบง่าย ตีความง่าย และเข้าใจผลลัพธ์ได้ชัดเจน ทำให้เหมาะสำหรับการนำไปใช้งานจริงในงานด้านการศึกษา เช่น การวางแผนการเรียน หรือระบบแนะนำพฤติกรรมสำหรับนักเรียนได้อย่างมีประสิทธิภาพ

**ดังนั้น Multiple Linear Regression คือโมเดลที่เหมาะสมที่สุด**

ข้อมูลเชิงลึก (Insight) และการนำไปต่อยอดหรือใช้ประโยชน์


จากการวิเคราะห์ข้อมูลโดยใช้เทคนิคทาง Data Science และ Machine Learning โครงงานนี้ช่วยค้นหาปัจจัยสำคัญและสะท้อนรูปแบบพฤติกรรมที่ส่งผลต่อผลการเรียน อย่างมีนัยสำคัญ ความเข้าใจเชิงลึกเหล่านี้สามารถนำไปใช้ปรับปรุงการเรียน การจัดการเวลาให้สอดคล้องกับเป้าหมายของนักเรียนแต่ละคน สำหรับการวางแผนการเรียนในอนาคตได้อย่างมีประสิทธิภาพ
โดยเราได้ค้นพบข้อมูลเชิงลึก ดังนี้ :
1. จำนวนที่ใช้อ่านหนังสือต่อวัน (study_hours_per_day) เป็นปัจจัยอันดับ 1 ที่มีผลต่อผลการเรียนแบบชัดเจนที่สุด เพราะมีคะแนน Feature Important ที่สูงกว่าตัวอื่นอย่างมาก แสดงว่าถ้าเพิ่มชั่วโมงอ่านหนังสือขึ้นทีละน้อย ผลการเรียนของนักเรียนจะดีขึ้นอย่างมีนัยสำคัญ
2. อัตราส่วนเวลาอ่านหนังสือ/เวลานอน (study_sleep_ratio) ถือเป็นปัจจัยที่สำคัญเป็นอันดับ 2 แสดงว่าไม่ใช่แค่ชั่วโมงอ่านหนังสือที่สำคัญ แต่ สมดุลระหว่างเวลานอนกับเวลาที่อ่านหนังสือ ก็สำคัญต่อ performance หากอ่านหนังสือหนักแต่พักผ่อนไม่เพียงพอจะทำให้คะแนนไม่ดี แต่การที่หาสมดุลทั้งสองค่าจะทำให้คะแนนดีขึ้น
3. สภาพจิตใจ (mental_health_rating) มีผลต่อผลการเรียนในระดับที่น้อยกว่า แต่ยังมีอิทธิพลอยู่บ้าง นักเรียนที่มีความเครียดจะทำให้คะแนนแย่ลง
4. การใช้ social media ไม่ได้ทำให้ผลการเรียนแย่ลง แต่การที่ผู้ปกครองเกิดความเข้าใจผิดว่าการใช้ social media ทำให้คะแนนแย่ลง คาดว่าเกิดจากนักเรียนที่เล่น social media อย่างหนักจนไม่มีเวลาอ่านหนังสือมากกว่าตามสมมติฐาน
5. attendance_percentage เปอร์เซ็นการเข้าเรียนไม่ได้ส่งผลต่อคะแนนมากนัก แต่การอ่านหนังสือนอกเวลากลับมีผลมากกว่า แสดงว่าการเรียนในห้องไม่ใช่ตัวแปรสำคัญที่สุดสำหรับผลการเรียนใน dataset นี้

และข้อเสนอการนำไปต่อยอด ดังนี้ :
1. ทำระบบ Personalized Study Recommendation แนะนำพัฒนาคะแนนสำหรับนักเรียนแต่ละคน เช่น ต้องเพิ่มชั่วโมงอ่านหนังสืออีกกี่ชั่วโมงเพื่อเพิ่มคะแนนตามเป้าหมาย ควรปรับการนอนเพื่อให้สมดุลขึ้นอย่างไร
2. ระบบ Predictive Early Warning เหมาะสำหรับคุณครู เพื่อคาดการณ์ว่ามีนักเรียนคนไหนเสี่ยงสอบตก
3. ทำ Dashboard เพื่อดูภาพรวมพฤติกรรมต่าง ๆ ว่ามีแนวโน้มอย่างไรและถ้าปรับค่าจะส่งผลต่อภาพรวมข้อมูลยังไงเพื่อนำไปปรับใช้ได้อย่างถูกต้อง
4. ใช้สร้าง Policy ให้กับสถาบันการศึกษา เพื่อปรับปรุงหลักสูตร ไม่ว่าจะเป็น การลดชั่วโมงเรียนเพื่อเพิ่ม self-learning, การออกแบบโปรแกรมสุขภาพจิต
